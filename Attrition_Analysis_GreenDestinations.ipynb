This project analyzes attrition at Green Destinations, a travel agency, to help the HR Director understand why employees are leaving.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\HP\Downloads\greendestination.csv")
df.head()
```

## ðŸ§¹ Data Cleaning
Before starting the analysis, we check for:
- Missing values
- Duplicate rows
- Data type consistency


```python
# ðŸ“Œ Basic Data Cleaning Checks

# 1. Check for missing values
missing_values = df.isnull().sum()
print("Missing Values in Each Column:\n", missing_values[missing_values > 0])

# 2. Check for duplicate rows
duplicate_count = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_count}")

# 3. Check data types
print("\nData types of columns:\n", df.dtypes)
```

## ðŸŽ¯ Objective
- Calculate the **attrition rate**.
- Understand how **Age**, **Monthly Income**, and **Years at Company** relate to attrition.


```python
# Calculate attrition rate
left_count = df[df['Attrition'] == 'Yes'].shape[0]
total_count = df.shape[0]
attrition_rate = round((left_count / total_count) * 100, 2)

sns.countplot(data=df, x='Attrition',hue = 'Attrition', palette='pastel',legend = False)
plt.title(f"Attrition Count (Rate: {attrition_rate}%)")
plt.show()
```

## ðŸ“ˆ Factor Analysis
Visualizing how key factors relate to attrition.


```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.set_palette("pastel")

sns.boxplot(data=df, x='Attrition', y='Age',hue ='Attrition', ax=axes[0])
axes[0].set_title('Age vs Attrition')

sns.boxplot(data=df, x='Attrition', y='YearsAtCompany', hue = 'Attrition', ax=axes[1])
axes[1].set_title('Years at Company vs Attrition')

sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', hue = 'Attrition', ax=axes[2])
axes[2].set_title('Monthly Income vs Attrition')

plt.tight_layout()
plt.show()
```

## âœ… Conclusion
- Attrition Rate is **16.12%**.
- Employees who leave tend to be **younger**, have **lower income**, and **shorter tenure**.
- Focus retention efforts on **newer and younger employees** with competitive pay.
